## Заполнение пропусков

Пропуски обнаружены в столбцах 'days_emploed' и 'total_income'.
Напишем функцию, которая вернет положительные значения. И заполним пропуски медианным значением. 

```python
df['total_income'].describe()
```

```python
def minus(value):    
    if value < 0:
        value *= -1
        return value
    else:
        return value
df['days_employed']=df['days_employed'].apply(minus)
```

```python
t_i_median = df['total_income'].median()
```

```python
d_e_median=df['days_employed'].median()
```

```python
df['total_income']=df['total_income'].fillna(t_i_median)
```

```python
df['days_employed']=df['days_employed'].fillna(d_e_median)
```

```python
df['children'] = df['children'].replace(-1, 1)
```

## Категоризация дохода.

На основании диапазонов, указанных в задании, создадим столбец 'total_income_category' с категориями.

```python
def cate(inc):
    if inc<=30000:
        return 'E'
    if 30001<=inc<=50000:
        return 'D'
    if 50001<=inc<=200000:
        return 'C'
    if 200001<=inc<=1000000:
        return 'B'
    if 1000001<=inc:
        return 'A'
df['total_income_category']=df['total_income'].apply(cate)
```

```python
df['total_income_category'].unique()
```

## Категоризация целей кредита.

Создадим функцию, которая на основании данных из столбца 'purpose' сформирует новый столбец "purpose_category", для упрощения возмем по три буквы из неоткатегоризированного столбца. Проведем проверку.

```python
def category(znach):
    if 'авт' in znach:
        return 'операции с автомобилем'
    if 'нед' in znach:
        return 'операции с недвижимостью'
    if 'жил' in znach:
        return 'операции с недвижимостью'
    if 'сва' in znach:
        return 'проведение свадьбы'
    if 'обр' in znach:
        return 'получение образования'
    else:
        return '1'
df['purpose_category']=df['purpose'].apply(category)
```

## Ответы на вопросы.

### Вопрос 1: Есть ли зависимость между количеством детей и возвратом кредита в срок?

```python
child_pivot=df.pivot_table(index=['children'], values=["debt"], aggfunc=['sum', 'count'])
child_pivot['prop']=child_pivot['sum']/child_pivot['count']*100
child_pivot
```

Без детей клиенты исправнее платят кредит. Для клиентов с пятью детьми выводы делать рано, тк выборка мальенькая.
Получается, что среднестатистическая семья попадает в категорию, не рекомендуемую для выдачи кредита...

### Вопрос 2: Есть ли зависимость между семейным положением и возвратом кредита в срок?

```python
df_fam=df.merge(newfam, on='family_status_id', how='left')
family_pivot=df_fam.pivot_table(index=['family_status'], values=["debt"], aggfunc=['sum', 'count'])
family_pivot['prop']=family_pivot['sum']/family_pivot['count']*100
family_pivot
```

Вдовцы(-вы) реже всех имеют просрочку по кредиту.
А неженатые и те, кто состоят в гражданском браке, чаще всех имеют просрочку по кредиту.

### Вопрос 3: Есть ли зависимость между уровнем дохода и возвратом кредита в срок?

```python
inc_pivot=df.pivot_table(index=['total_income_category'], values=["debt"], aggfunc=['sum', 'count'])
inc_pivot['prop']=inc_pivot['sum']/inc_pivot['count']*100
inc_pivot
```

Вывод сделаем по информативным категориям (B,C). Чем выше доход, тем реже просрочки по кредиту. Правильнее было бы выбрать медиану не по всему столбцу, а привязать к категориям клиентов. Так, 2174 пропуска распределились бы более плавно, а не в середине из-за малеького разнообразия выборки.


### Вопрос 4: Как разные цели кредита влияют на его возврат в срок?

```python
pur_pivot=df.pivot_table(index=['purpose_category'], values=["debt"], aggfunc=['sum', 'count'])
pur_pivot['prop']=pur_pivot['sum']/pur_pivot['count']*100
pur_pivot
```

Меньше всего просрочек (в процентном соотношении) при покупке недвижимости. 
При покупке автомобиля и кредита на образование чаще встречаются невыплаты.
Ну а при проведении свадьбы доля просроченных выплат составляет около 8%.
